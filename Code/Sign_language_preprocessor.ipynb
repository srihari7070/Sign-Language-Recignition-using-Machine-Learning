```python
#import the libraries
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten
from keras.datasets import mnist
from keras.utils import to_categorical
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
from tqdm import tqdm
import random
import pickle


# In[27]:

#DATADIR = "C:/Users/Srihari Ananthan/Desktop/SLRuML/DataSets/First"
#DATADIR = "C:/Users/Srihari Ananthan/Desktop/SLRuML/DataSets/Second"
#DATADIR = "C:/Users/Srihari Ananthan/Desktop/SLRuML/DataSets/Third/training_set"
DATADIR = "C:/Users/Srihari Ananthan/Desktop/SLRuML/DataSets/hk"
#DATADIR = "C:/Users/Srihari Ananthan/Desktop/SLRuML/Digital numbers"

CATEGORIES = ["E","F","Y","M","D","C","H"]
#CATEGORIES = ["0","1","2","3","4"]

for category in CATEGORIES:  
    path = os.path.join(DATADIR,category)  # create path
    for img in os.listdir(path):  # iterate over each image
        img_array = cv2.imread(os.path.join(path,img))  # convert to array
        plt.imshow(img_array, cmap='gray')  # graph it
        plt.show()  # display!

        break  # we just want one for now so break
    break  #...and one more!


# In[28]:


IMG_SIZE = 100
new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(new_array, cmap='gray')
plt.show()


# In[29]:


training_data = []

def create_training_data():
    for category in CATEGORIES:  

        path = os.path.join(DATADIR,category)  
        class_num = CATEGORIES.index(category)  # get the classification

        for img in tqdm(os.listdir(path)):  # iterate over each image 
            try:
                img_array = cv2.imread(os.path.join(path,img))  # convert to array
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))  # resize to normalize data size
                training_data.append([new_array, class_num])  # add this to our training_data
            except Exception as e:  # in the interest in keeping the output clean...
                pass
            #except OSError as e:
            #    print("OSErrroBad img most likely", e, os.path.join(path,img))
            #except Exception as e:
            #    print("general exception", e, os.path.join(path,img))

create_training_data()

print(len(training_data))


# In[30]:


random.shuffle(training_data)
#for sample in training_data[:10]:
    #print(sample[1])


# In[31]:


X = []
y = []

for features,label in training_data:
    X.append(features)
    y.append(label)

   # print(X[0].reshape(-1, IMG_SIZE, IMG_SIZE, 1))

X = np.array(X).reshape(-3, IMG_SIZE, IMG_SIZE, 3)
y = np.array(y)

print("efvef x", X.shape)


# In[32]:


pickle_out = open("x_asl1.pickle","wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y_asl1.pickle","wb")
pickle.dump(y, pickle_out)
pickle_out.close()


# In[ ]:
```
